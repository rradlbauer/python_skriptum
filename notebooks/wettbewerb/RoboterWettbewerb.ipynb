# Roboter Wettbewerb

## Cyberbotics

Cyberbotics bietet die Software Webots an, mit welcher es möglich ist, das Verhalten von Robotern zu simulieren. Diese simulierten Roboter können mit Python programmiert werden - eine gute Möglichkeit für einen Wettbewerb.

## Software

Installiere die Webots-Applikation:
https://github.com/cyberbotics/webots/releases/download/R2025a/webots-R2025a_setup.exe

Spiele für den Einstieg dieses Tutorial durch:
https://cyberbotics.com/doc/guide/tutorial-1-your-first-simulation-in-webots
Das sollte reichen, damit du eine Ahnung davon hast, wie Webots funktioniert.

Um Grundlagen zu erlernen, wie man einen Roboter Controller programmiert, ist dieses Tutorial zu empfehlen:
https://cyberbotics.com/doc/guide/tutorial-4-more-about-controllers
Für den Wettbewerb wird genau der Roboter aus diesem Tutorial verwendet.

## Vorbereitung

Lade das Projekt [wettbewerb](https://downgit.github.io/#/home?url=https://github.com/rradlbauer/python_skriptum/tree/master/notebooks/wettbewerb)
herunter und entpacke die ZIP-Datei.

Öffne mit Webots die darin enthaltene Datei `wettbewerb.wbt` im Unterverzeichnis `worlds`.
Diese "Welt" enthält eine Arena mit einem E-Puck Roboter und einer Holzkiste.
Im Verzeichnis `world` ist ein Python-Skript mit dem Namen `single-box.py` enthalten.
Wenn du dieses Skript ausführst (`py single-box.py` in einer Eingabeaufforderung), dann wird die "Welt" neu erstellt,
und die Holzbox erscheint - wenn du die "Welt" neu ladest - an einer zufälligen Stelle.

Beim Öffnen von `wettbewerb.wbt` wird auch gleich das Python-Skript `wettbewerb.py` geöffnet,
welches den Controller-Code vom E-Puck enthält. Dieser tut im Moment nichts.

## Wettbewerb 1

Erstelle einen Controller, welcher möglichst schnell die Kiste findet und die Koordinaten möglichst genau ausgibt.
(Die Ausgabe erfolgt ganz normal mit `print(...)`).
Lass den Controller mit verschiedenen zufälligen Positionen der Holzkiste laufen,
um zu sehen, ob der Roboter die Kiste auch jedes Mal findet.

### Wertung

Dein Controller wird nach folgenden Kriterien bewertet:
* Genauigkeit der Koordinaten (öffne `wettbewerb.wbt` mit einem Texteditor, damit du siehst, wie die Koordinaten definiert werden.)
* Zeit, die der Roboter braucht, um die Kiste zu finden
